Since we use ReLU as activation function, and as a result we get some non-covex non-linear piecewise function, that has many local minima. And in some cases, the different set of parameters gives the same function.

For example, imagine that some hidden unit is inactive because of x < 0 => z = 0, and changing parameters of it will not change the function at all.

The least squares loss for this shallow ReLU network is nonconvex and therefore does not have a unique minimum.
Multiple different parameter combinations can yield the same output function (due to hidden unit symmetry and ReLU saturation).
Gradient descent finds a local or global minimum, but not necessarily the unique one.

First of all, if activation function is linear - than the number of linear regions will be zero.
